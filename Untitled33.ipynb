```python
import pandas as pd
m=[[4,5],[6,5],[7,8]]
n=[[1,2],[3,4],[4,5]]
c=["A",'B']
l=pd.DataFrame(m,columns=c)
h=pd.DataFrame(n,columns=c)
result=(l,h)
k=pd.concat(result)
print(l)
print()
print(h)
print()
print(k)
```

```python
import pandas as pd
m=[[4,5],[6,5],[7,8]]
n=[[1,2],[3,4],[4,5]]
c=["A","B"]

l=pd.DataFrame(m,columns=c,)
h=pd.DataFrame(n,columns=c)
result=(l,h)
k=pd.concat(result,ignore_index=1)
print(l)
print()
print(h)
print()
print(k)
```

```python
import pandas as pd
m=[[4,5],[6,5],[7,8]]
n=[[1,2],[3,4],[4,5]]
c=["A","B"]

l=pd.DataFrame(m,columns=c,)
h=pd.DataFrame(n,columns=c)
result=(l,h)
k=pd.concat(result,axis=1)
print(l)
print()
print(h)
print()
print(k)
```
