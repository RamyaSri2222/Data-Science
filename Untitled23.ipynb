```python
import matplotlib.pyplot as plt
x=[10,25,35,40]
y=[5,6,7,8]
m=plt.bar(x,y)
print(m)
```

```python
import matplotlib.pyplot as plt
x=[12,13,14,18,19]
y=[15,18,34,45,80]
plt.xlabel("Grade")
plt.ylabel("Marks")
plt.title("Student age & marks")
n=plt.barh(x,y)
print(n)
```

```python
import matplotlib.pyplot as plt
x=[15,30,40,50,60]

m=plt.hist(x,bins=20)
print(m)
```

```python
import matplotlib.pyplot as plt
x=[10,20,30]
y=[15,30,40]
k=[12,14,16,18,20]
c=['y','r','g']
plt.title("Data analysis")
m=plt.scatter(x,y,label=k,color=c)
print(m)
```
