```python
import matplotlib.pyplot as plt
import pandas as pd


data = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Math': [85, 90, 78, 92],
    'Science': [88, 76, 95, 89],
    'English': [90, 85, 80, 86]
})


data['Total'] = data[['Math', 'Science', 'English']].sum(axis=1)


plt.figure(figsize=(8, 5))
plt.bar(data['Name'], data['Total'], color='skyblue', edgecolor='black')
plt.title('Total Marks of Students')
plt.xlabel('Student Name')
plt.ylabel('Total Marks')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()


subject_avg = data[['Math', 'Science', 'English']].mean()

plt.figure(figsize=(6, 6))
plt.pie(subject_avg, labels=subject_avg.index, autopct='%1.1f%%', startangle=90)
plt.title('Subject-wise Average Marks')
plt.legend(subject_avg.index, loc='best')
plt.show()
```

```python
import matplotlib.pyplot as plt


months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
sales = [25000, 27000, 30000, 28000, 32000, 35000, 37000, 39000, 41000, 43000, 45000, 47000]

plt.figure(figsize=(8, 5))
plt.plot(months, sales, marker='o', color='blue', linestyle='-', linewidth=2, label='Sales (Line)')
plt.title('Monthly Sales Data')
plt.xlabel('Months')
plt.ylabel('Sales Amount')
plt.grid(True)
plt.legend()
plt.show()


plt.figure(figsize=(8, 5))
plt.bar(months, sales, color='orange', label='Sales (Bar)')
plt.title('Monthly Sales Data')
plt.xlabel('Months')
plt.ylabel('Sales Amount')
plt.grid(axis='y')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


movies = ['Inception', 'Avatar', 'Titanic', 'Interstellar', 'Joker']
imdb = [8.8, 7.8, 7.9, 8.6, 8.5]
rotten = [87, 82, 89, 93, 68]
metacritic = [74, 83, 75, 74, 59]


x = np.arange(len(movies))
width = 0.25


plt.bar(x - width, imdb, width, label='IMDB')
plt.bar(x, rotten, width, label='Rotten Tomatoes')
plt.bar(x + width, metacritic, width, label='Metacritic')


plt.title('Movie Ratings Comparison')
plt.xlabel('Movies')
plt.ylabel('Ratings')
plt.xticks(x, movies, rotation=30)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt


math = [88, 92, 79, 93, 85, 91, 87, 95, 77, 89]
science = [75, 80, 72, 78, 85, 90, 88, 84, 82, 76]
english = [70, 68, 75, 80, 72, 74, 77, 73, 79, 71]
history = [65, 60, 70, 68, 72, 75, 78, 74, 71, 69]


data = [math, science, english, history]

plt.figure(figsize=(7, 5))
plt.boxplot(data, patch_artist=True, notch=True,
            boxprops=dict(facecolor='lightblue'),
            medianprops=dict(color='red'))

plt.title('Distribution of Exam Scores Across Subjects')
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.xticks([1, 2, 3, 4], ['Math', 'Science', 'English', 'History'])
plt.grid(axis='y', linestyle='--', alpha=0.7)


plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


ages = [18, 20, 22, 25, 27, 28, 30, 32, 33, 35, 36, 38, 40, 42, 44, 45, 46, 48, 50, 52, 54, 55, 56, 58, 60]


plt.figure(figsize=(8, 5))
count, bins, patches = plt.hist(ages, bins=8, color='skyblue', edgecolor='black', alpha=0.7, density=True, label='Age Distribution')


mean = np.mean(ages)
std = np.std(ages)
x = np.linspace(min(ages), max(ages), 100)
y = (1 / (std * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std)**2)
plt.plot(x, y, color='red', linewidth=2, label='Density Curve')


plt.title('Age Distribution of Event Attendees')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend()
plt.grid(alpha=0.5)


plt.show()

```
